```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
csv_path1 = "raw_data/clinicaltrial_data.csv"
csv_path2 = "raw_data/mouse_drug_data.csv"

clin_df = pd.read_csv(csv_path1)
mouse_df = pd.read_csv(csv_path2)
```

```python
list(clin_df)
```

```python
cm_df = pd.merge(clin_df, mouse_df, how='outer', on='Mouse ID')
cm_df.head()
```

```python
# Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
tvol_df=cm_df.groupby(['Drug','Timepoint'])
tvol_df=tvol_df.mean()
tvol_df=tvol_df.drop(columns="Metastatic Sites")

#Creating a scatter plot that shows the Metastatic Sites
ms_df=cm_df.groupby(['Drug','Timepoint'])
ms_df=ms_df.mean()
ms_df=ms_df.drop(columns="Tumor Volume (mm3)")


#Creating a scatter plot that shows the Metastatic Sites
ap_df=cm_df.groupby(['Drug','Timepoint']).count()["Mouse ID"]
#We understood survival rate as the count of all the different types of Mouse Id's and how they decrease over the timepoint per drug
ap_df = ap_df.reset_index()

#ap_df.head()
# ms_df.head()
tvol_df.head()
```

```python
#switch index to column names as indiciated in example
swap_df = pd.pivot_table(tvol_df, values="Tumor Volume (mm3)", index="Timepoint",columns="Drug")
swap_df.head()

swapms_df = pd.pivot_table(ms_df, values="Metastatic Sites", index="Timepoint",columns="Drug")
# swapms_df

sr_df = pd.pivot_table(ap_df, values="Mouse ID", index="Timepoint", columns="Drug")
# sr_df
```

```python
plt.errorbar(swap_df.index,swap_df["Capomulin"], marker="o", color="r", ls='--', linewidth=.5)
plt.errorbar(swap_df.index,swap_df["Infubinol"], marker = "^", color ="b", linestyle='--', linewidth=.5)
plt.errorbar(swap_df.index,swap_df["Ketapril"], marker = "s", color ="g", linestyle='--', linewidth=.5)
plt.errorbar(swap_df.index,swap_df["Placebo"], marker = "D", color ="black", linestyle='--', linewidth=.5)
plt.grid()

plt.legend(loc='upper left')

plt.title("Tumor Reponse to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
plt.errorbar(swapms_df.index,swapms_df["Capomulin"], marker="o", color="r", ls='--', linewidth=.5)
plt.errorbar(swapms_df.index,swapms_df["Infubinol"], marker = "^", color ="b", linestyle='--', linewidth=.5)
plt.errorbar(swapms_df.index,swapms_df["Ketapril"], marker = "s", color ="g", linestyle='--', linewidth=.5)
plt.errorbar(swapms_df.index,swapms_df["Placebo"], marker = "D", color ="black", linestyle='--', linewidth=.5)

plt.grid()

plt.legend(loc='upper left')
plt.title("Metastatic Spread Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Met. Sites")
plt.show()
```

```python
plt.errorbar(sr_df.index,sr_df["Capomulin"]*100/sr_df["Capomulin"][0], marker="o", color="r", ls='--', linewidth=.5)
plt.errorbar(sr_df.index,sr_df["Infubinol"]*100/sr_df["Infubinol"][0], marker = "^", color ="b", linestyle='--', linewidth=.5)
plt.errorbar(sr_df.index,sr_df["Ketapril"]*100/sr_df["Ketapril"][0], marker = "s", color ="g", linestyle='--', linewidth=.5)
plt.errorbar(sr_df.index,sr_df["Placebo"]*100/sr_df["Placebo"][0], marker = "D", color ="black", linestyle='--', linewidth=.5)

plt.grid()

plt.legend(loc='lower left')
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate(%)")

plt.show()
```
